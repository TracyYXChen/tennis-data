```python
import pandas as pd
import os


PROCESSED_DIR = 'processed'
firstYearMatch = os.path.join(PROCESSED_DIR, 'firstYearMatches.csv')
matchDf = pd.read_csv(firstYearMatch, sep=',')
matchDf
```

```python
playerProfile = os.path.join(PROCESSED_DIR, 'processedList.csv')
playerDf = pd.read_csv(playerProfile, sep=',')
playerDf
```

## first year win rates vs. highest ranking

```python
def calcFirstYearWinRate():
    winRateDict = {}
    allPlayers = list(playerDf['pid'])
    for py in allPlayers:
        wl = list(matchDf[matchDf['pid'] == py]['wl'])
        w = wl.count('W')
        if len(wl) == 0:
            winRateDict[py] = -1
        else:    
            winRateDict[py] = round(w / len(wl), 3)  
    return winRateDict

FYRATE = 'firstYearWinRate'
winRateDict = calcFirstYearWinRate()
playerDf[FYRATE] = playerDf.apply(lambda row: winRateDict[row['pid']], axis=1) 
#drop players without 1st year match history
playerDf = playerDf[playerDf[FYRATE] != -1]
playerDf.sort_values(by=[FYRATE], ascending=False)
playerDf

# highest ranking rank_single_ch
```

```python
# group ranking by groups
RK_GRP = 'rankingGroup'
TP100 = 'top100'
TP50 = 'top50'
TP30 = 'top30'
TP10 = 'top10'
HIGHEST_RK = 'rank_single_ch'
rankDict = {1: 'top 10', 2: 'top 30', 3: 'top 50', 4: 'top 100', 5: 'others'}

def calcRkGroup():
    grpDict = {}
    top100Dict = {}
    top50Dict = {}
    top30Dict = {}
    top10Dict = {}
    allPlayers = list(playerDf['pid'])
    for py in allPlayers:
        highestRk = int(playerDf[playerDf['pid'] == py][HIGHEST_RK])
        rkGrp = ''
        top100 = False
        top50 = False
        top30 = False
        top10 = False
        if highestRk > 100:
            rkGrp = rankDict[5]
        elif highestRk > 50:
            rkGrp = rankDict[4]
            top100 = True
        elif highestRk > 30:
            rkGrp = rankDict[3]
            top100 = True
            top50 = True
        elif highestRk > 10:
            rkGrp = rankDict[2]
            top100 = True
            top50 = True
            top30 = True
        elif highestRk > 0:
            rkGrp = rankDict[1]
            top100 = True
            top50 = True
            top30 = True
            top10 = True
        else:
            print('%s, rank: %s'%(py, highestRk))
        grpDict[py] = rkGrp
        top100Dict[py] = top100
        top50Dict[py] = top50
        top30Dict[py] = top30
        top10Dict[py] = top10
    return [grpDict, top100Dict, top50Dict, top30Dict, top10Dict]

[grpDict, top100Dict,  top50Dict, top30Dict, top10Dict] = calcRkGroup()
playerDf[RK_GRP] = playerDf.apply(lambda row: grpDict[row['pid']], axis=1) 
playerDf[TP100] = playerDf.apply(lambda row: top100Dict[row['pid']], axis=1) 
playerDf[TP50] = playerDf.apply(lambda row: top50Dict[row['pid']], axis=1) 
playerDf[TP30] = playerDf.apply(lambda row: top30Dict[row['pid']], axis=1) 
playerDf[TP10] = playerDf.apply(lambda row: top10Dict[row['pid']], axis=1) 
playerDf
          
    
```

```python
# rankingGroup vs. firstYearWinRate
# winRate, height, rankingGroup
import altair as alt
import random

#remove height < 1m
playerDf = playerDf[playerDf['height'] > 100]
#manually add a jitter
playerDf['height_jittered'] = playerDf.apply(lambda row: row['height'] + random.uniform(-1,1), axis=1) 
#remove old data
playerDf = playerDf[playerDf['pro_year'] > 2000]
playerDf = playerDf[playerDf[FYRATE] > 0.6]
# too crowded

tp_ls = [TP100, TP50, TP30, TP10]


alt.Chart(playerDf, title="First-year win rates for players turned pro after 2000").mark_circle(size=60).encode(
        alt.X('height_jittered',
        scale=alt.Scale(zero=False)
        ),
        alt.Y(FYRATE,
        scale=alt.Scale(zero=False)
        ),
        color=TP30,
        tooltip=['first_name', 'last_name', HIGHEST_RK, 'height', 'pro_year', FYRATE]
).interactive()

```

```python
# try aggregation
import statistics

getFYRate = lambda grpName : statistics.mean(list(playerDf[playerDf[RK_GRP] == grpName][FYRATE]))
getSTDRate = lambda grpName: statistics.stdev(list(playerDf[playerDf[RK_GRP] == grpName][FYRATE]))
meanRateDict = {'top 10': getFYRate('top 10'), 'top 30': getFYRate('top 30'), 
                'top 50': getFYRate('top 50'), 'top 100': getFYRate('top 100'), 'others': getFYRate('others')}

MEAN_RATE = 'meanWinRate'
STD_RATE = 'stdRate'
grpLs = ['top 10', 'top 30', 'top 50', 'top 100', 'others']
source = pd.DataFrame({
    RK_GRP: grpLs,
    MEAN_RATE: list(map(getFYRate, grpLs)),
    STD_RATE: list(map(getSTDRate, grpLs))
})
```

```python
error_bars = alt.Chart(playerDf, title="First-year win rates for players turned pro after 2000").mark_errorbar(extent='stdev').encode(
  x=alt.X('%s:Q'%FYRATE, scale=alt.Scale(zero=False)),
  y=alt.Y('%s:N'%RK_GRP, title='ranking', sort='x')
)

points = alt.Chart(playerDf).mark_point(filled=True, color='black').encode(
  x=alt.X('%s:Q'%FYRATE, aggregate='mean'),
  y=alt.Y('%s:N'%RK_GRP, title='ranking', sort='x')
)

points + error_bars 
```
