```python
import pandas as pd
import os
import altair as alt
```

```python
RAW_DIR = "raw"
wta2022 = os.path.join(RAW_DIR, "wta_matches_2022.csv")
wta2023 = os.path.join(RAW_DIR, "wta_matches_2023.csv")
playerDf2022 = pd.read_csv(wta2022, sep=',')
playerDf2023 = pd.read_csv(wta2023, sep=',')
print(playerDf2023.columns)
print(playerDf2022.columns)
```

```python
queen2022 = playerDf2022[(playerDf2022['winner_name'] == "Qinwen Zheng") | (playerDf2022['loser_name'] == "Qinwen Zheng")]
#queen2023 = playerDf2023[(playerDf2023['winner_name'] == "Qinwen Zheng") | (playerDf2023['loser_name'] == "Qinwen Zheng")]
#queenDf = pd.concat([queen2022, queen2023])
queenDf = queen2022
queenDf = queenDf[(queenDf['surface'] == "Hard") | (queenDf['surface'] == "Clay")]
queenDf['won'] = queenDf.apply(lambda row: row['winner_name'] == 'Qinwen Zheng', axis=1)
queenDf
```

```python
alt.Chart(queenDf, title="first serve in vs. first serve win").mark_circle(size=60).encode(
        alt.X('w_1stIn',
        scale=alt.Scale(zero=True)
        ),
        alt.Y('w_1stWon',
        scale=alt.Scale(zero=True)
        ),
        color= 'won', 
        tooltip=['tourney_name', 'tourney_date', 'score', 'winner_name', 'loser_name', 'w_1stIn', 'w_1stWon', 'l_1stIn', 'l_1stWon', 'w_2ndWon', 'l_2ndWon' ]
).interactive()
```
